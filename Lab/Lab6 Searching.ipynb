Q1  Given an array of n elements, find two elements in the array such that their sum is

equal to K. The two elements can be the same element. Once a pair of elements is

found, the program can be terminated. The function prototype is given below:

`def dual_search(A, size, K, dual_index):`

#A: The input array of integers.

#size: The size of the array.

#K: The target sum.

#dual_index: A list to store the indices of the two elements.

#return true if pars are found, otherwise return false

An example input is:

A = [3, 1, 7, 4, 5, 9]

K = 8

The output is:

Pair found at indices: [0, 4]

Elements: 3 + 5 = 8

```python
def dual_search(A, size, K, dual_index):
    dul_index=[]
    for i in range(size):
        for j in range(size):
            if A[i]+A[j]==K:
                dual_index.append(i)
                dual_index.append(i)
                return dual_index
    return []
```

Q2  Given a sorted array of n elements (you can use merge sort (Lab 5) to sort

the array), find two elements in the array such that their sum is equal to K.

The two elements can be the same element. Once a pair of elements are

found, the program can be terminated. The results may be different from

the results of Question 1. Your algorithm for searching the pair in the sorted

array must run in 𝑂(𝑛) time complexity. The function prototype is given

below:

`def dual_sorted_search(A, size, K, dual_index): `

#A: The input array of integers.

#size: The size of the array.

#K: The target sum.

#dual_index: A list to store the indices of the two elements.

#return true if pars are found, otherwise return false

An example input is:

A = [9, 1, 4, 3, 7, 5]

K = 8

The output is:

Pair found at indices: [0, 4]

Elements: 1 + 7 = 8

```python
def dual_sorted_search(A, size, K, dual_index):
    left,right=0,size-1
    while left<=right:
        if A[left]+A[right]==K:
            dual_index.append(left)
            dual_index.append(right)
            return dual_index
        elif A[left]+A[right]>K:
            right-=1
        else:
            left+=1
    return []
```
