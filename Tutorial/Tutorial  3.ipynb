### 2. Level order traversal

```python
def leve_order_traversal(root):
    if not root:
        return
    queue=[root]
    result=[]
    while queue:
        time=len(queue)
        current_level_nodes=[]
        for _ in range(time):
            node=queue.pop(0)
            current_level_nodes.append(node.val)
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)
        result.append(current_level_nodes)
    return result
```


### 3. Preorder Iterative: Preorder traversal using stack


```python
def preOrderIterative(root):
    if not root:
        return
    stack=[root]
    result=[]
    while stack:
        node=stack.pop()
        result.append(node.val)
        if node.right:
            stack.append(node.right)
        if node.left:
            stack.append(node.left)
```


### 4. Find the max depth of a tree

```python
def maxDepth(node):
    if not node:
        return -1
    return 1+max(maxDepth(node.right),maxDepth(node.left))
```
