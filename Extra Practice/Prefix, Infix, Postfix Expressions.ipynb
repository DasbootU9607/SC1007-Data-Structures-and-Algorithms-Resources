
### 1. Preparation (given)


```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class Stack:
    def __init__(self):
        self.top = None
        self.size = 0
    def isEmpty(self): return self.size == 0
    def push(self, data):
        new_node = Node(data)
        new_node.next = self.top
        self.top = new_node
        self.size += 1
    def pop(self):
        if self.isEmpty(): raise IndexError("Pop from empty stack")
        val = self.top.data
        self.top = self.top.next
        self.size -= 1
        return val
    def peek(self):
        if self.isEmpty(): raise IndexError("Peek from empty stack")
        return self.top.data

PRECEDENCE = {'+': 1, '-': 1, '*': 2, '/': 2, '**': 3}
```

### Question 1 Infix to Postfix

```python
def infix_to_postfix(expr_list):
    op_stack=Stack()
    result=Stack()
    for item in expr_list:
        # TODO: 实现你的逻辑
        # 1. 如果是操作数 -> 加入 result
        # 2. 如果是 "(" -> 入栈
        # 3. 如果是 ")" -> 弹栈直到遇到 "("
        # 4. 如果是运算符 -> 比较优先级并决定是否弹栈，最后入栈f
        if item not in PRECEDENCE and item not in "()":
            result.push(item)
        elif item=="(":
            op_stack.push(item)
        elif item==")":
            while op_stack.peek!="(" and not op_stack.isEmpty():
                result.push(op_stack.pop())
            op_stack.pop()
        else:
            while not op_stack.isEmpty and op_stack.peek()!="(":
                top=op_stack.peek()
                if PRECEDENCE[top]>PRECEDENCE[item] or (PRECEDENCE[top]==PRECEDENCE[item] and item!="**"):
                    result.push(op_stack.pop())
                else:
                    break
            op_stack.push(item)
    while not op_stack.isEmpty():
        result.push(op_stack.pop())
    return result
```

### Question 2 Postfix to Prefix

```python
def postfix_to_prefix(expr_list):
    s = Stack()
    result=Stack()
    for item in expr_list:
        if item not in PRECEDENCE:
            # TODO: 操作数入栈
            s.push(item)
        else:
            # TODO:
            # 1. 弹出两个操作数 (注意顺序：先弹 A，再弹 B)
            # 2. 拼接成新字符串: "item B A" (中间带空格)
            # 3. 压回栈
            op2=s.pop()
            op1=s.pop()
            new_exp=item+op1+op2
            s.push(new_exp)
    result.push(s.pop())
    return result
```

### Question 3 Prefix to Postfix

```python
def prefix_to_postfix(expr_list):
    s = Stack()
    # 提示：使用 reversed(expr_list) 来实现从右往左扫描
    result=Stack()
    for item in reversed(expr_list):
        if item not in PRECEDENCE:
            # TODO: 操作数入栈
            s.push(item)
        else:
            # TODO:
            # 1. 弹出两个操作数 (先弹 A，再弹 B)
            # 2. 拼接成新字符串: "A B item"
            # 3. 压回栈
            op1=s.pop()
            op2=s.pop()
            new_exp=op1+op2+item
            s.push(new_exp)
    result.push(s.pop())
    return result
```
